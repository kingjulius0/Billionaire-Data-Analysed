A quick look at previous Billionaires list to ascertain the total number of Self-made billionaires out of the 475 magnates and the top 15 Self-made in the world 🌏

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data = pd.read_csv('/content/Billionaires Statistics Dataset - Billionaires Statistics Dataset (1).csv')
```

```python
Data
```

Quick statistical summary


```python
Data.describe()
```

Various data types in view

```python
Data.info()
```

```python
Data.dropna(inplace=True)
```

Data cleaning: drop duplicates if any

```python
Data.drop_duplicates(inplace=True)
```

Limit data to 300 columns

```python
Data = Data.drop(Data.index[300:])
```

```python
Data
```

Top 10 Billionaires (self-made or not) 😊

```python
top_10_richest = Data.nlargest(10, 'finalWorth')
top_10_richest['firstName']
```

```python
top_10_richest_with_networth = top_10_richest[['firstName', 'finalWorth']]
top_10_richest_with_networth
```

Graphical visualisation of the 10 richest

```python
plot = sns.barplot(x='firstName', y='finalWorth', data=top_10_richest)
```

15 top SELF-MADE Billionaires with final worth (networth)

```python
Selfmade_billionaires = Data[Data['selfMade'] == True]
Selfmade_billionaires = Selfmade_billionaires.sort_values(by='finalWorth', ascending=False)
Selfmade_billionaires[['firstName', 'finalWorth']].head(15)
```

202 out of 475 billionairs are self-made

```python
Selfmade_billionaires['firstName'].count()
```

Visualization of the top 15 Self-made billionaires

```python
Selfmade_billionaires_plot = plt.figure(figsize=(10, 6))
plt.bar(x=Selfmade_billionaires.head(15)['firstName'], height=Selfmade_billionaires.head(15)['finalWorth']) ,
plt.xlabel('Name'),
plt.ylabel('Neworth'), plt.title('Top 15 Selfmade Billionaires'),
plt.xticks(rotation=90),

plt.show()
```
